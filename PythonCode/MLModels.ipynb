```python
# import the required libraries
import os
import csv
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)
```

```python

## Read the training and validation datasets
train =  pd.read_csv("../Data/Output/TrainingData.csv")
val = pd.read_csv("../Data/Output/ValidationData.csv")

## Check the data looks fine
# Understanding basic information about the data
print("Number of rows in train data = ", train.shape[0])
print("Number of columns in train data = ", train.shape[1])
print("\n")
print(" *** Sample Carrier Data ******")
display(train.head(5))

print("\n")
print("----------------------------------------------------------")
print("\n")


print("Number of rows in Validation Data = ", val.shape[0])
print("Number of columns in Validation Data = ", val.shape[1])
print("\n")
print("** Sample Validation Data **")
val.head()
```

```python
## Create a Training and Validation dataset with only the Commodities Columns
train_X_onlyCommodities = train.iloc[:, 19:50]
train_Y_onlyCommodities = train[['VAN', 'REFR', 'FLAT']]

val_X_onlyCommodities = val.iloc[:, 19:50]
val_Y_onlyCommodities = val[['VAN', 'REFR', 'FLAT']]

#### Create Training and Validation datasets with Number of Tractors and Trailers also
train_X_tractrail = train.iloc[:, np.r_[6:8,19:50]]
train_Y_tractrail = train[['VAN', 'REFR', 'FLAT']]

val_X_tractrail = val.iloc[:, np.r_[6:8,19:50]]
val_Y_tractrail = val[['VAN', 'REFR', 'FLAT']]
```

### One vs. All Classifiers

```python
### One vs. All Classifier -- with Logistic Regression

from sklearn.linear_model import LogisticRegression
from sklearn.pipeline import Pipeline
from sklearn import metrics
from sklearn.metrics import accuracy_score
from sklearn.multiclass import OneVsRestClassifier
```

```python
labels = list(train_Y_onlyCommodities.columns.values)

# Using pipeline for applying logistic regression and one vs rest classifier
LogReg_pipeline = Pipeline([
                ('clf', OneVsRestClassifier(LogisticRegression(solver='sag'), n_jobs=-1)),
            ])
for label in labels:
    print('**Processing {} Equipment Label...**'.format(label))
    
    # Training logistic regression model on train data
    LogReg_pipeline.fit(train_X_onlyCommodities, train_Y_onlyCommodities[label])
    
    # calculating test accuracy
    prediction = LogReg_pipeline.predict(val_X_onlyCommodities)
    print('Validation accuracy is {}'.format(accuracy_score(val_Y_onlyCommodities[label], prediction)))
    print("hamming loss :",metrics.hamming_loss(val_Y_onlyCommodities[label],prediction))
    print("Precision recall report :\n",metrics.classification_report(val_Y_onlyCommodities[label], prediction))
    print("\n")
```

```python
#### Build the model after including the number of trailer and tractor information -- this is a proxy for the Carrier Size
labels = list(train_Y_tractrail.columns.values)

# Using pipeline for applying logistic regression and one vs rest classifier
LogReg_pipeline = Pipeline([
                ('clf', OneVsRestClassifier(LogisticRegression(solver='sag'), n_jobs=-1)),
            ])
for label in labels:
    print('**Processing {} Equipment Label...**'.format(label))
    
    # Training logistic regression model on train data
    LogReg_pipeline.fit(train_X_tractrail, train_Y_onlyCommodities[label])
    
    # calculating test accuracy
    prediction = LogReg_pipeline.predict(val_X_tractrail)
    print('Validation accuracy is {}'.format(accuracy_score(val_Y_tractrail[label], prediction)))
    print("hamming loss :",metrics.hamming_loss(val_Y_tractrail[label],prediction))
    print("Precision recall report :\n",metrics.classification_report(val_Y_tractrail[label], prediction))
    print("\n")
```

Including the number of trailers and tractors seems to slighly improve the performance metric for the VAN predictions

### Binary Relevance Classifier

```python
# using binary relevance
from skmultilearn.problem_transform import BinaryRelevance
from sklearn.naive_bayes import GaussianNB
from sklearn import metrics
from sklearn.metrics import f1_score,precision_score,recall_score, hamming_loss

# initialize binary relevance multi-label classifier
# with a gaussian naive bayes base classifier
classifier = BinaryRelevance(GaussianNB())

# train
classifier.fit(train_X_onlyCommodities, train_Y_onlyCommodities)
# predict
predictions = classifier.predict(val_X_onlyCommodities)
# accuracy

#print("Accuracy = ",accuracy_score(val_Y_onlyCommodities,predictions))
#print("Hamming loss ",hamming_loss(val_Y_onlyCommodities,predictions))
#print("Precision ",precision_score(val_Y_onlyCommodities,predictions))


print("accuracy :",metrics.accuracy_score(val_Y_onlyCommodities,predictions))
print("macro f1 score :",metrics.f1_score(val_Y_onlyCommodities, predictions, average = 'macro'))
print("micro f1 scoore :",metrics.f1_score(val_Y_onlyCommodities, predictions, average = 'micro'))
print("hamming loss :",metrics.hamming_loss(val_Y_onlyCommodities,predictions))
print("Precision recall report :\n",metrics.classification_report(val_Y_onlyCommodities, predictions))
```
